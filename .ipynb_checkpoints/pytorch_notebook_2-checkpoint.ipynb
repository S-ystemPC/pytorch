# Gradient Calculation

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

y = x+2
print(y)
z = y*y*2
z = z.mean()
print(z)

z.backward() # dz/dx
print(x.grad)
```

# Vector and Jacobian product gives a scalar output (Chain Rule)
# When Vector is not provided see the error

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

y = x+2
print(y)
z = y*y*2
# z = z.mean()
print(z)

z.backward() # dz/dx
print(x.grad)
```

# Passing Vector to it

```python
import torch

x = torch.randn(3, requires_grad=True)
print(x)

y = x+2
print(y)
z = y*y*2
# z = z.mean()
print(z)

v = torch.tensor([0.1, 1.0, 0.001], dtype=torch.float32)
z.backward(v) # dz/dx
print(x.grad)
```

# Prevent PyTorch from tracking history and calculating "grad_fn" attribute

```python
x = torch.randn(3, requires_grad=True)
print(x)

# x.requires_grad_(False)
# x.detach()
# with torch.no_grad():

# 1st option
x.requires_grad_(False) # Whenever a function has a trailing underscore in PyTorch then this means that it will modify our variable in place.
print(x)

# 2nd option
y = x.detach()
print(y)

# last option
with torch.no_grad():
    z = x+2
    print(z)
```

# Dummy Training example

```python
import torch

weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
    model_output = (weights*3).sum()

    model_output.backward()

    print(weights.grad)

    weights.grad.zero_()
```

# PyTorch built-in Optimizer

```python
import torch

weights = torch.ones(4, requires_grad=True)

optimizer = torch.optim.SGD(weights, lr=0.01)
optimizer.step()
optimizer.zero_grad()
```
