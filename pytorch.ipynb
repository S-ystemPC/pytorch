# Some Basic Operations

```python
# Addition
a1 = torch.rand(2,2)
a2 = torch.rand(2,2)
print(a1)
print(a2)
a3 = a1 + a2
print(a3)

# Or
a4 = torch.add(a1,a2)
print(a4)
```

```python
# Subtraction
a4 = torch.rand(2,2)
a5 = torch.rand(2,2)

a6 = a4 - a5
print(a6)

# Or
a7 = torch.sub(a4, a5)
print(a7)
```

```python
# Multiplication
a8 = torch.rand(2,2)
a9 = torch.rand(2,2)

a10 = a8 * a9
print(a10)

# Or
a11 = torch.mul(a8, a9)
print(a11)
```

```python
# Division
a12 = torch.rand(2,2)
a13 = torch.rand(2,2)

a14 = a12/a13
print(a14)

# Or
a15 = torch.div(a12,a13)
print(a15)
```

# In-place operation

```python
# Addition
# a1 = torch.rand(2,2)
# a2 = torch.rand(2,2)
a1.add_(a2)
print(a1)
```

```python
# Subtraction
# a4 = torch.rand(2,2)
# a5 = torch.rand(2,2)

a4.sub_(a5)
print(a4)
```

```python
# Multiplication
# a8 = torch.rand(2,2)
# a9 = torch.rand(2,2)
a8.mul_(a9)
print(a8)
```

```python
# Division
# a12 = torch.rand(2,2)
# a13 = torch.rand(2,2)
a12.div_(a13)
print(a12)
```

# Slicing Operation

```python
b1 = torch.rand(5,3)
print(b1)
# For all columns
print(b1[:,0])
```

```python
# For all rows
print(b1[0,:])
```

```python
print(b1[1,1])
```

```python
# Can be used when only one element in the tensor, to get the value actually stored in the tensor
print(b1[1,1].item())
```

# Reshaping a tensor

```python
b2 = torch.rand(4,4)
print(b2)
b3 = b2.view(16)
print(b3)
```

```python
# if don't want to determine the size use this style

b4 = torch.rand(4,4)
print(b4)
b5 = b4.view(-1,8)
# (-1,8), here 8 states the no. of element in the row and rest the dimension pytorch will determine
print(b5)
print(b5.size())
```
