# 2.1 Empty Tensors

```python
import torch

# 1 inside small braces is like scaler value
x1 = torch.empty(1)
print(x1)
```

```python
import torch

# 3 inside small braces is like 1D vector with three elements
x2 = torch.empty(3)
print(x2)
```

```python
# two values inside small braces means 2D vector

x3 = torch.empty(2, 2)

print(x3)
```

```python
# 3 dimension

x4 = torch.empty(2, 2, 3)
print(x4)
```

# Tensor with random values

```python
x6 = torch.rand(2,2)
print(x6)
```

# Tensor with all zeroes

```python
x7 = torch.zeros(2,2)
print(x7)
```

# Tensor with all ones

```python
x8 = torch.ones(2,2)
print(x8)
```

# Specific data type

```python
x9 = torch.ones(2,2)
print(x9.dtype)
```

```python
x10 = torch.ones(2,2, dtype=torch.int)
print(x10.dtype)
```

```python
x11 = torch.ones(2,2, dtype=torch.double)
print(x11.dtype)
```

```python
x12 = torch.ones(2,2, dtype=torch.float16)
print(x12.dtype)
```

# Size

```python
x13 = torch.ones(2,2, dtype=torch.int)
print(x13.size())
```

# Construct a Tensor from data

```python
x14 = torch.tensor([2.5, 0.1])
print(x14)
```
